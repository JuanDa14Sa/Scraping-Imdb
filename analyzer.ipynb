```python
import networkx as nx
import json
import matplotlib.pyplot as plt
import collections
```

```python
# Load the data
data = json.load(open('top250movies.json', 'r'))

movie_graph = nx.DiGraph()  # Create a graph

for movie in data:
    movie_graph.add_node(movie['Title'], title=movie['Title'], year=movie['Year'], link=movie['Link'], type='movie')
    for actor in movie['Actors'][0:5]:
        movie_graph.add_node(actor, title=actor, type='actor')    
        movie_graph.add_edge(actor, movie['Title'])
```

```python
movie_graph.order()
```

```python
sorted(list(movie_graph.out_degree), key=lambda x: x[1], reverse=True) 
```

```python
color_dict = {'movie': 'red', 'actor': 'blue'} 
nx.draw(movie_graph, with_labels=False, node_size=1, node_color=[color_dict[movie_graph.nodes[node]['type']] for node in movie_graph.nodes], width=0.1, arrowsize=0.1)   
```
